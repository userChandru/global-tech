```python
%sql

CREATE SCHEMA IF NOT EXISTS 01_global_tech_bronze.raw_tables;
```

```python
csv_files = [
    "accounts.csv",
    "company.csv", 
    "departments.csv", 
    "employee.csv", 
    "general ledgers.csv", 
    "payroll.csv"
]

for file in csv_files:
    table_name = file.replace(".csv", "_raw")
    df = spark.read.format("csv").options(header=True, inferSchema=True).load(f"s3://global-tech-raw/raw_files/{file}")
    df.write.format("delta").mode("overwrite").saveAsTable(f"01_global_tech_bronze.raw_tables.{table_name.replace(" ", "_")}")
```
