```python
from pyspark.sql.functions import trim, col, monotonically_increasing_id, split
```

```python
accounts_df = spark.table("01_global_tech_bronze.raw_tables.accounts_raw")

transformed_accounts_df = (
    accounts_df
    .withColumn("account_code", split(col("account_code"), "-")[1].cast("int"))    
    .withColumn("account_name", trim(col("account_name")))
    .withColumn("account_type", trim(col("account_type")))
    .withColumn("category", trim(col("category")))
    .withColumn("account_sk", monotonically_increasing_id())
    .withColumnRenamed("is_active", "account_is_active")
)

transformed_accounts_df.write.format("delta").mode("overwrite").saveAsTable("02_global_tech_silver.transformed_tables.transformed_accounts")

# display(transformed_accounts_df)
```
