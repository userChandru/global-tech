```python
from pyspark.sql.functions import trim, col, monotonically_increasing_id
```

```python
dept_df = spark.table("01_global_tech_bronze.raw_tables.departments_raw")

transformed_departments = (
    dept_df
    .withColumn("department_code", trim(col("department_code")))
    .withColumn("cost_center", trim(col("cost_center")))    
    .withColumn("department_name", trim(col("department_name")))
    .withColumn("department_sk", monotonically_increasing_id())
    .select(
        "department_sk", "department_id", "department_code", 
        "department_name", "cost_center"
    )
)
transformed_departments.write.format("delta").mode("overwrite").saveAsTable("02_global_tech_silver.transformed_tables.transformed_departments")

# display(transformed_departments)
```
